```python
from requests_html import HTMLSession
from csv import writer
import pandas as pd
```

### I was planning to download weather data of all taluk in india but direct scrapping large amount of weather data from google sucks. It stops continueosly throwing error after aproximately 200 over data collected and there is another problem you can't continuosly use it, it deosn't work. 
### So I only collected all district weather data in my state(karnataka) only.

```python

district = ('Bagalkot', 'Bangalore', 'Bangalore Rural', 'Belgaum', 'Bellary', 'Bidar', 'Bijapur', 'Chamarajanagar', 'Chikkaballapura', 'Chikmagalur', 'Chitradurga', 'Dakshina Kannada', 'Davanagere', 'Dharwad', 'Gadag', 'Gulbarga', 'Hassan', 'Haveri', 'Kodagu', 'Kolar', 'Koppal', 'Mandya', 'Mysore', 'Raichur', 'Ramanagara', 'Shimoga', 'Tumkur', 'Udupi', 'Uttara Kannada', 'Yadgir')
Names = len(district)
Names
for i in range(Names):
    try:
        session = HTMLSession()
        query = district[i]
        print(i)
        url = f'https://www.google.com/search?q=weather+{query}'
        request = session.get(url, headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.115 Safari/537.36'})
        location = request.html.find('div#wob_loc', first =True).text
        date_time = request.html.find('div#wob_dts', first =True).text
        temperature = request.html.find('span#wob_tm', first =True).text
        unit = request.html.find('<span class="wob_t" style="display:inline" aria-label="°Celsius" aria-disabled="true" role="button">°C</span>', first =True).text
        description = request.html.find('span#wob_dc', first =True).text
        print(location, date_time, temperature, unit, description)
        List=[location, date_time, temperature, unit, description]
        with open('all_distict_Karnataka_weather_data.csv', 'a') as f_object:
            writer_object = writer(f_object)
            writer_object.writerow(List)
            f_object.close()
        
    except:
        continue


```
